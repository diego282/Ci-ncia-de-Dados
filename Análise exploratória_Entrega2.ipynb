```python
#Função para contar quantidade de respostas negativas
def conta_respostas_nao(coluna):
  cont_2 = 0
  for i in coluna:
    if not math.isnan(i):
        if (i == 2):
          cont_2 = cont_2 + 1
  return cont_2
```

```python
#Função para contar quantidade de respostas positivas
def conta_respostas_sim(coluna):
  cont_1 = 0
  for i in coluna:
    if not math.isnan(i):
        if (int(i) == 1):
          cont_1 = cont_1 + 1
  return cont_1
```

```python
#Função para contar quantidade de respostas que não foram respondidas
def conta_respostas_vazias(coluna):
  cont_3 = 0
  for i in coluna:
    if (math.isnan(i) == True) :
          cont_3 = cont_3 + 1
  return cont_3
```

```python
#Depressão
coluna = 'Q092'
cont_sim = conta_respostas_sim(pd.to_numeric(df[coluna]))
cont_nao = conta_respostas_nao(pd.to_numeric(df[coluna]))
cont_vazio = conta_respostas_vazias(pd.to_numeric(df[coluna]))
print("Depressão: \n")
print("Respostas positivas: " , cont_sim )
print("Respostas negativas: " , cont_nao )
print("Respostas vazias: " , cont_vazio )
```

```python
#Função que calcular a quantidade de homens e mulheres que responderam sim a pergunta se tem depressão
def calcula_quantidade_depressao():
    countM = 0
    countF = 0
    for i in range(0, len(df['Q092'])):
      if not math.isnan(df['Q092'][i]) and int(df['Q092'][i]) == 1: # Se não é NaN E Se tem depressão
          if not math.isnan(df['C006'][i]): # Se não é NaN
              if int(df['C006'][i]) == 1: # se é homem
                countM += 1
              if int(df['C006'][i]) == 2: # se é mulher
                countF += 1
    return [countM, countF]

print("Quantidade de respostas positivas por genero: \n")
print(calcula_quantidade_depressao())
```

```python
def separa_homens_nao(): # Homens que disseram que não tem depressão
    homens_nao_total=[]
    for i in range(0, len(df['Q092'])):
        if not math.isnan(df['C006'][i]) and int(df['C006'][i]) == 1 and not math.isnan(df['Q092'][i]) and int(df['Q092'][i]) == 2:#se é homem e respondeu que não possui depressão
            homens_nao_total.append(i)
    return homens_nao_total
```
