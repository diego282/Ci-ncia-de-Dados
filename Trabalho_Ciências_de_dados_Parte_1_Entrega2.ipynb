```python
import pandas as pd
import re
import math
```

```python
widths = []
headers = []
```

```python
#Função para definir o tamanho dos dados em cada coluna da base
def get_width(line):
    s = line.split("\t") #Separa os dados do arquivo pela tabulação
    r = re.search(r"\d", s[2]) # Realiza uma procura por dígito , selecionando a segunda posição do vetor s

    return int(r.group())#Retorna o tamanho dos dados de cada coluna da base a ser utilizada
```

```python
#Função para retornar o código das variáveis da base
def get_header(line):
  s = line.split("\t")
  return s[1]
```

```python
#Abre o arquivo em modo leitura e insere todos os dados na variável f
with open('input_PNS_2019.txt', 'r', encoding='utf-8') as f:
    for line in f:
       widths.append(get_width(line.strip())) #adiciona no vetor widths o tamanho correspondente a cada coluna
       headers.append(get_header(line.strip()))#adiciona no vetor headers o código da variável correspondente a cada coluna
```

```python
#Atribui ao df (dataframe) os dados separados com seus respectivos tamanhos e códigos das variáveis referentes às colunas
df = pd.read_fwf('PNS_2019.txt', widths=widths, header=None, dtype=str)
df.to_csv("base_dados.csv", sep=";", encoding="utf-8", index=False)
```

```python
#Função para contar quantidade de respostas positivas
def conta_respostas_sim(coluna):
  cont_1 = 0
  for i in coluna:
    if not math.isnan(i): 
        if (int(i) == 1 ):
          cont_1 = cont_1 + 1
  return cont_1
```

```python
#Função para contar quantidade de respostas negativas
def conta_respostas_nao(coluna):
  cont_2 = 0
  for i in coluna:
    if not math.isnan(i):
        if (int(i) == 2 ):
          cont_2 = cont_2 + 1
  return cont_2
```

```python
#Função para contar quantidade de respostas que não responderam
def conta_respostas_vazias(coluna):
  cont_3 = 0
  for i in coluna:
    if (math.isnan(i) == True) :
          cont_3 = cont_3 + 1
  return cont_3
```
