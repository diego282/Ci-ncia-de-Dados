```python
#Função para contar quantidade de respostas negativas
def conta_respostas_nao(coluna):
  cont_2 = 0
  for i in coluna:
    if not math.isnan(i):
        if (i == 2):
          cont_2 = cont_2 + 1
  return cont_2
```

```python
#Função para contar quantidade de respostas positivas
def conta_respostas_sim(coluna):
  cont_1 = 0
  for i in coluna:
    if not math.isnan(i):
        if (int(i) == 1):
          cont_1 = cont_1 + 1
  return cont_1
```

```python
#Função para contar quantidade de respostas que não responderam
def conta_respostas_vazias(coluna):
  cont_3 = 0
  for i in coluna:
    if (math.isnan(i) == True) :
          cont_3 = cont_3 + 1
  return cont_3
```

```python
#Depressão
coluna = 'Q092'
cont_sim = conta_respostas_sim(pd.to_numeric(df[coluna]))
cont_nao = conta_respostas_nao(pd.to_numeric(df[coluna]))
cont_vazio = conta_respostas_vazias(pd.to_numeric(df[coluna]))
print("Depressão: \n")
print("Respostas positivas: " , cont_sim )
print("Respostas negativas: " , cont_nao )
print("Respostas vazias: " , cont_vazio )
```

```python
def plot_stacked_plot_percentages(class_list, df_p, add_x_val):
    
    '''Esta função adiciona a porcentagem em cada barra do gráfico
    class_list: Contém três parâmetros, a contagem de dada classe/barra, o total para cada classe/barra, o objeto ax da classe/barra
    df_p      : Parte do dataframe com o objeto de cada retângulo
    add_x_val : Valor do eixo x para centralizar o texto
    '''
    
    for class_val in class_list:
        
        percentage = (class_val[0]/class_val[1])*100
        height = class_val[2].get_height() 
        df_p.axes.text(class_val[2].get_x()+add_x_val, height,"{:1.1f}%".format(percentage))
```

```python
def calculate_stacked_plot_percentages(df_plot, df_p):
    df_plot_sum = df_plot.sum(axis=1)
    class_sum = np.array([df_plot_sum.iloc[int(i/2)] if i%2==0 else df_plot_sum.iloc[int((i-1)/2)] for i in range(2*len(df_plot_sum)) ]).reshape((len(df_plot_sum),2)).T.flatten()
    class_list = zip(np.roll(df_plot.to_numpy().T.flatten(), len(df_plot)), class_sum, list(df_p.axes.patches))

    return class_list
```

```python
sns.set(rc = {'figure.figsize':(15,8)})

# Encontrando as faixas
df['age_bin'] = pd.cut(df['C008'], 3, precision=0)

# Cálculo a contagem de variável categórica 'survived' para diferentes faixas etárias
df_plot = df.groupby([ 'Q092', 'age_bin']).size().reset_index().pivot(columns='Q092', index='age_bin', values=0)
print(df_plot)

# Gráfico de barra empilhada
df_p = df_plot.plot(kind='bar', stacked=True)

# Defina os rótulos
df_p.set_xlabel("Grupos de idade", fontsize = 20)
df_p.set_ylabel("Qtde", fontsize = 20)

class_list = calculate_stacked_plot_percentages(df_plot, df_p)

plot_stacked_plot_percentages(class_list, df_p, 0.12)
```
